A continuación, se presenta el desarrollo paso a paso, ordenado y explicado, reconstruyendo las ecuaciones que las diapositivas resumen.

---
### 🔹 1. Ecuación de movimiento relativo
Partimos de la ecuación diferencial del movimiento relativo entre dos cuerpos sometidos a gravitación mutua:
$$
\ddot{\mathbf{r}} = -\frac{\mu}{r^3}\mathbf{r}
$$
donde:
- $\mathbf{r}$ es el vector de posición relativa,
- $r = |\mathbf{r}|$,
- $\mu = G(m_1 + m_2)$ es el parámetro gravitacional estándar.

---
### 🔹 2. Primera cuadratura: Momento Angular Relativo Específico (M.A.R.E.)
Multiplicamos vectorialmente la ecuación de movimiento por $\mathbf{r}$ por la izquierda:
$$
\mathbf{r} \times \ddot{\mathbf{r}} = -\frac{\mu}{r^3}(\mathbf{r} \times \mathbf{r}) = \mathbf{0}
$$
Como $\frac{d}{dt}(\mathbf{r} \times \dot{\mathbf{r}}) = \dot{\mathbf{r}} \times \dot{\mathbf{r}} + \mathbf{r} \times \ddot{\mathbf{r}} = \mathbf{0}$, se deduce que:
$$
\mathbf{h} \equiv \mathbf{r} \times \mathbf{v} = \text{cte.}
$$
Este vector constante $\mathbf{h}$ es el **M.A.R.E.** (Momento Angular Relativo Específico). Su conservación implica:
- El movimiento es plano (perpendicular a $\mathbf{h}$).
- Se satisface la segunda ley de Kepler (velocidad areolar constante).

---
### 🔹 3. Segunda cuadratura: Energía Relativa Específica (E.R.E.)
Multiplicamos escalarmente la ecuación de movimiento por $\dot{\mathbf{r}}$:
$$
\dot{\mathbf{r}} \cdot \ddot{\mathbf{r}} = -\frac{\mu}{r^3}\mathbf{r} \cdot \dot{\mathbf{r}}
$$
Usando identidades de derivación:
$$
\frac{d}{dt}\left(\frac{v^2}{2}\right) = -\mu \frac{d}{dt}\left(\frac{1}{r}\right) \quad \Rightarrow \quad \frac{d}{dt}\left(\frac{v^2}{2} - \frac{\mu}{r}\right) = 0
$$
Por tanto:
$$
\mathcal{E} \equiv \frac{v^2}{2} - \frac{\mu}{r} = \text{cte.}
$$
Esta es la **E.R.E.** (Energía Relativa Específica). Su valor determina el tipo de cónica: $\mathcal{E}<0$ (elipse), $\mathcal{E}=0$ (parábola), $\mathcal{E}>0$ (hipérbola).

---
### 🔹 4. Tercera cuadratura: Vector de Excentricidad (Vector de Laplace-Runge-Lenz)
Para encontrar una tercera integral vectorial, cruzamos la ecuación de movimiento con $\mathbf{h}$:
$$
\ddot{\mathbf{r}} \times \mathbf{h} = -\frac{\mu}{r^3}\mathbf{r} \times (\mathbf{r} \times \mathbf{v})
$$
Aplicamos la **regla BAC-CAB** ($\mathbf{A}\times(\mathbf{B}\times\mathbf{C}) = \mathbf{B}(\mathbf{A}\cdot\mathbf{C}) - \mathbf{C}(\mathbf{A}\cdot\mathbf{B})$):
$$
\mathbf{r} \times (\mathbf{r} \times \mathbf{v}) = \mathbf{r}(\mathbf{r}\cdot\mathbf{v}) - \mathbf{v}(\mathbf{r}\cdot\mathbf{r}) = r\dot{r}\mathbf{r} - r^2\mathbf{v}
$$
Sustituyendo:
$$
\ddot{\mathbf{r}} \times \mathbf{h} = -\frac{\mu}{r^3}(r\dot{r}\mathbf{r} - r^2\mathbf{v}) = \mu\left(\frac{\mathbf{v}}{r} - \frac{\dot{r}\mathbf{r}}{r^2}\right)
$$
Notamos que el lado derecho es exactamente la derivada temporal de $\frac{\mathbf{r}}{r}$:
$$
\frac{d}{dt}\left(\frac{\mathbf{r}}{r}\right) = \frac{\dot{\mathbf{r}}}{r} - \frac{\mathbf{r}\dot{r}}{r^2}
$$
Por tanto:
$$
\frac{d}{dt}(\dot{\mathbf{r}} \times \mathbf{h}) = \frac{d}{dt}\left(\mu\frac{\mathbf{r}}{r}\right) \quad \Rightarrow \quad \frac{d}{dt}\left(\frac{\dot{\mathbf{r}} \times \mathbf{h}}{\mu} - \frac{\mathbf{r}}{r}\right) = \mathbf{0}
$$
Definimos así el **vector de excentricidad** (o vector de Laplace):
$$
\boxed{\mathbf{e} = \frac{\mathbf{v} \times \mathbf{h}}{\mu} - \frac{\mathbf{r}}{r} = \text{cte.}}
$$

---
### 🔹 5. Interpretación geométrica del vector $\mathbf{e}$
- **Dirección**: $\mathbf{e}$ yace en el plano orbital (es combinación lineal de $\mathbf{r}$ y $\mathbf{v}$) y apunta siempre hacia el **periastro** (punto de mínima distancia).
- **Magnitud**: Su norma $e = |\mathbf{e}|$ es la **excentricidad** de la trayectoria. Se relaciona con las integrales escalares mediante:
$$
e^2 = 1 + \frac{2\mathcal{E}h^2}{\mu^2}
$$
Esta relación se obtiene calculando $\mathbf{e}\cdot\mathbf{e}$ y sustituyendo $\mathbf{v}\cdot\mathbf{v}$ y $\mathbf{h}\cdot\mathbf{h}$ en función de $\mathcal{E}$ y $h$.

---
### 🔹 6. Ecuación de la trayectoria en coordenadas polares
Proyectamos $\mathbf{e}$ sobre $\mathbf{r}$ usando el producto escalar:
$$
\mathbf{e} \cdot \mathbf{r} = \left(\frac{\mathbf{v} \times \mathbf{h}}{\mu} - \frac{\mathbf{r}}{r}\right) \cdot \mathbf{r}
$$
Como $(\mathbf{v} \times \mathbf{h}) \cdot \mathbf{r} = \mathbf{h} \cdot (\mathbf{r} \times \mathbf{v}) = \mathbf{h} \cdot \mathbf{h} = h^2$, y $\frac{\mathbf{r}}{r}\cdot\mathbf{r} = r$, obtenemos:
$$
e r \cos\theta = \frac{h^2}{\mu} - r
$$
donde $\theta$ es el ángulo entre $\mathbf{e}$ y $\mathbf{r}$ (verdadera anomalía). Despejando $r$:
$$
r(1 + e\cos\theta) = \frac{h^2}{\mu}
$$
Definiendo el **parámetro semilatus rectum** $p \equiv \frac{h^2}{\mu}$, llegamos a la **ecuación polar de la trayectoria**:
$$
\boxed{r(\theta) = \frac{p}{1 + e\cos\theta}}
$$
Esta es la ecuación general de una **sección cónica** con foco en el origen, confirmando que las órbitas del problema de dos cuerpos son elipses, parábolas o hipérbolas según el valor de $e$ (y por ende, de $\mathcal{E}$).

---
### 🔹 7. Resumen de cuadraturas obtenidas
| Integral | Expresión | Tipo | Significado físico |
|----------|-----------|------|---------------------|
| M.A.R.E. | $\mathbf{h} = \mathbf{r} \times \mathbf{v}$ | Vectorial (3 componentes) | Plano orbital + conservación de área |
| E.R.E.   | $\mathcal{E} = \frac{v^2}{2} - \frac{\mu}{r}$ | Escalar (1) | Tipo de órbita (cerrada/abierta) |
| Vector de Laplace | $\mathbf{e} = \frac{\mathbf{v} \times \mathbf{h}}{\mu} - \frac{\mathbf{r}}{r}$ | Vectorial (3 componentes, pero 2 independientes por estar en el plano) | Orientación y forma de la cónica |

En total se obtienen **6 cuadraturas** (3 de $\mathbf{h}$, 1 de $\mathcal{E}$, 2 independientes de $\mathbf{e}$), que permiten integrar completamente el problema relativo sin recurrir a cuadraturas adicionales.

## 🔹 1. Ecuación del movimiento y cuadraturas
Partimos de la ecuación diferencial del movimiento relativo:
$$\ddot{\vec{r}} = -\frac{\mu}{r^3}\vec{r}$$
donde $\mu = G(m_1+m_2)$ y $r=|\vec{r}|$.

**Step 1. Primera cuadratura (M.A.R.E.)**
$$\vec{r} \times \ddot{\vec{r}} = \vec{0} \quad \Rightarrow \quad \frac{d}{dt}(\vec{r} \times \vec{v}) = \vec{0}$$
$$\vec{h} \equiv \vec{r} \times \vec{v} = \text{cte.}$$
*Explicación:* Conservación del momento angular específico. El movimiento es plano y satisface $r^2\dot{\theta}=h$.

**Step 2. Segunda cuadratura (E.R.E.)**
$$\dot{\vec{r}} \cdot \ddot{\vec{r}} = -\frac{\mu}{r^3}\vec{r}\cdot\dot{\vec{r}} \quad \Rightarrow \quad \frac{d}{dt}\left(\frac{v^2}{2} - \frac{\mu}{r}\right) = 0$$
$$\mathcal{E} \equiv \frac{v^2}{2} - \frac{\mu}{r} = \text{cte.}$$
*Explicación:* Conservación de la energía específica. Determina si la órbita es cerrada ($\mathcal{E}<0$) o abierta ($\mathcal{E}\geq 0$).

**Step 3. Tercera cuadratura (Vector de Laplace)**
$$\ddot{\vec{r}} \times \vec{h} = -\frac{\mu}{r^3}\vec{r} \times (\vec{r} \times \vec{v}) = \mu\frac{d}{dt}\left(\frac{\vec{r}}{r}\right)$$
Integrando:
$$\vec{e} \equiv \frac{\vec{v} \times \vec{h}}{\mu} - \frac{\vec{r}}{r} = \text{cte.}$$
*Explicación:* Vector constante en el plano orbital. Su dirección apunta al periastro y su norma $e=|\vec{e}|$ es la excentricidad.

---

## 🔹 2. Ecuación de la trayectoria en coordenadas polares
**Step 1. Proyección escalar**
$$\vec{e} \cdot \vec{r} = \left(\frac{\vec{v} \times \vec{h}}{\mu} - \frac{\vec{r}}{r}\right) \cdot \vec{r}$$
Usando $(\vec{v} \times \vec{h})\cdot\vec{r} = \vec{h}\cdot(\vec{r}\times\vec{v}) = h^2$:
$$e r \cos\theta = \frac{h^2}{\mu} - r$$

**Step 2. Despeje de $r(\theta)$**
Definiendo el semilatus rectum $p \equiv \frac{h^2}{\mu}$:
$$\boxed{r(\theta) = \frac{p}{1 + e\cos\theta}}$$
*Explicación:* Ecuación polar general de una cónica con foco en el origen.

---

## 🔹 3. Definición geométrica (Arquímedes)
**Step 1. Propiedad foco-directriz**
Para cualquier punto $P$ de la cónica, foco $F$ y directriz $D$:
$$\frac{PF}{PD} = e = \text{cte.}$$

**Step 2. Configuración cartesiana**
Colocamos el foco en el origen $(0,0)$ y la directriz vertical en $x=x_d$. Para $P(x,y)$:
$$PF = \sqrt{x^2+y^2}, \quad PD = |x_d - x|$$
$$\sqrt{x^2+y^2} = e|x_d - x|$$

**Step 3. Ecuación cartesiana respecto al foco**
Elevando al cuadrado y reorganizando:
$$y^2 = (e^2-1)x^2 + 2e^2 x_d x - e^2 x_d^2$$
Introduciendo $p = e x_d$ (parámetro semilatus rectum):
$$\boxed{y^2 = 2px + (e^2-1)x^2}$$
*Explicación:* Forma general de la cónica con origen en el foco.

---

## 🔹 4. Traslación de coordenadas
### 4.1 Traslación al apside (vértice)
El periastro está en $x = q = \frac{p}{1+e}$. Definimos $x = x_a + q$:
$$y_a^2 = 2p(x_a+q) + (e^2-1)(x_a+q)^2$$
Simplificando y usando $p = q(1+e)$:
$$y_a^2 = 2p x_a + (e^2-1)x_a^2$$
*Explicación:* La forma se mantiene, pero el origen se sitúa en el vértice de la órbita.

### 4.2 Traslación al centro de la cónica
Buscamos eliminar el término lineal en $x$. Hacemos $x_a = x_c + a$, donde $a$ es la distancia centro-apside:
$$a = \frac{p}{|1-e^2|}$$
Sustituyendo y completando cuadrados:
$$y_c^2 + (1-e^2)x_c^2 = \frac{p^2}{1-e^2}$$
Dividiendo por el término independiente:
$$\boxed{\frac{x_c^2}{a^2} + \frac{y_c^2}{b^2} = 1} \quad \text{con} \quad b^2 = a^2(1-e^2)$$
*Explicación:* Ecuación canónica centrada en el origen geométrico de la cónica.

---

## 🔹 5. Propiedades geométricas según el tipo de cónica

| Propiedad | Elipse ($0\leq e < 1$) | Hipérbola ($e > 1$) | Parábola ($e = 1$) |
|-----------|------------------------|---------------------|-------------------|
| **Semieje mayor** | $a = \frac{p}{1-e^2}$ | $a = \frac{p}{e^2-1}$ | $a \to \infty$ |
| **Semieje menor** | $b = a\sqrt{1-e^2}$ | $b = a\sqrt{e^2-1}$ | No aplica |
| **Distancia centro-foco** | $c = ea$ | $c = ea$ | No aplica |
| **Relación geométrica** | $a^2 = b^2 + c^2$ | $c^2 = a^2 + b^2$ | $y^2 = 4fx$ |
| **Asíntotas** | No tiene | $y_c = \pm \frac{b}{a}x_c$ | No tiene |

**Step 1. Elipse**
- Ecuación: $\frac{x_c^2}{a^2} + \frac{y_c^2}{b^2} = 1$
- El foco está desplazado $c=ea$ del centro. La curva es cerrada y acotada.

**Step 2. Hipérbola**
- Ecuación: $\frac{x_c^2}{a^2} - \frac{y_c^2}{b^2} = 1$
- Asíntotas rectas con pendiente $\pm b/a$. La curva es abierta y tiende a las asíntotas cuando $x_c \to \infty$.

**Step 3. Parábola**
- Límite $e\to 1 \Rightarrow a\to\infty$. La directriz se aleja al infinito.
- Ecuación desde el vértice: $y^2 = 4fx$ (con $f$ distancia focal).

# Desarrollo Analítico para Determinar $e^2$

**Objetivo:** Expresar la magnitud al cuadrado del vector de excentricidad, $e^2 = \vec{e} \cdot \vec{e}$, en función de las integrales primeras del problema: la energía relativa específica $\mathcal{E}$ y el momento angular específico $h = |\vec{h}|$.

---

### 🔹 Paso 1: Definición del vector de excentricidad
Partimos del vector de Laplace (constante de movimiento):
$$\vec{e} = \frac{\vec{v} \times \vec{h}}{\mu} - \frac{\vec{r}}{r}$$
*Explicación:* Es una combinación lineal de vectores en el plano orbital. Su norma representa la excentricidad geométrica de la trayectoria.

---

### 🔹 Paso 2: Producto escalar consigo mismo
Calculamos $e^2 = \vec{e} \cdot \vec{e}$ expandiendo el binomio:
$$e^2 = \left( \frac{\vec{v} \times \vec{h}}{\mu} - \frac{\vec{r}}{r} \right) \cdot \left( \frac{\vec{v} \times \vec{h}}{\mu} - \frac{\vec{r}}{r} \right)$$
$$e^2 = \frac{1}{\mu^2}(\vec{v} \times \vec{h}) \cdot (\vec{v} \times \vec{h}) - \frac{2}{\mu}\frac{\vec{r}}{r} \cdot (\vec{v} \times \vec{h}) + \frac{\vec{r}}{r} \cdot \frac{\vec{r}}{r}$$
*Explicación:* Desarrollo algebraico directo del producto escalar. A continuación, analizaremos cada término por separado.

---

### 🔹 Paso 3: Tercer término (unidad)
$$\frac{\vec{r}}{r} \cdot \frac{\vec{r}}{r} = \frac{\vec{r} \cdot \vec{r}}{r^2} = \frac{r^2}{r^2} = 1$$
*Explicación:* El vector unitario radial $\hat{r}$ tiene, por definición, una norma igual a $1$.

---

### 🔹 Paso 4: Segundo término (producto mixto)
$$\frac{\vec{r}}{r} \cdot (\vec{v} \times \vec{h}) = \frac{1}{r} \, \vec{h} \cdot (\vec{r} \times \vec{v}) = \frac{1}{r} \, \vec{h} \cdot \vec{h} = \frac{h^2}{r}$$
*Explicación:* Se utiliza la propiedad cíclica del producto triple escalar y la definición $\vec{h} = \vec{r} \times \vec{v}$. El término completo en la ecuación de $e^2$ queda como: $-\dfrac{2h^2}{\mu r}$.

---

### 🔹 Paso 5: Primer término (módulo de producto vectorial doble)
Usamos la identidad vectorial $|\vec{A} \times \vec{B}|^2 = A^2 B^2 - (\vec{A} \cdot \vec{B})^2$:
$$(\vec{v} \times \vec{h}) \cdot (\vec{v} \times \vec{h}) = v^2 h^2 - (\vec{v} \cdot \vec{h})^2$$
Como $\vec{h} = \vec{r} \times \vec{v}$, entonces $\vec{v} \cdot \vec{h} = \vec{v} \cdot (\vec{r} \times \vec{v}) = 0$ (producto triple con dos vectores idénticos). Por tanto:
$$(\vec{v} \times \vec{h}) \cdot (\vec{v} \times \vec{h}) = v^2 h^2$$
*Explicación:* El término se simplifica drásticamente porque el vector velocidad siempre es perpendicular al vector de momento angular específico.

---

### 🔹 Paso 6: Ensamblaje y sustitución de la energía
Reunimos los tres resultados obtenidos:
$$e^2 = \frac{v^2 h^2}{\mu^2} - \frac{2h^2}{\mu r} + 1$$
Factorizamos el término $\dfrac{h^2}{\mu^2}$:
$$e^2 = 1 + \frac{h^2}{\mu^2} \left( v^2 - \frac{2\mu}{r} \right)$$
Recordamos la definición de energía relativa específica: $\mathcal{E} = \dfrac{v^2}{2} - \dfrac{\mu}{r}$. Multiplicando esta expresión por $2$:
$$2\mathcal{E} = v^2 - \frac{2\mu}{r}$$
Sustituimos finalmente en la expresión de la excentricidad:
$$\boxed{e^2 = 1 + \frac{2\mathcal{E}h^2}{\mu^2}}$$

---

### 🔹 Interpretación física
Esta relación es fundamental en mecánica celeste, ya que conecta la geometría de la sección cónica con las constantes físicas del sistema:

* **Si $\mathcal{E} < 0$**: $e^2 < 1 \rightarrow$ **Elipse** (órbita ligada).
* **Si $\mathcal{E} = 0$**: $e^2 = 1 \rightarrow$ **Parábola** (trayectoria de escape límite).
* **Si $\mathcal{E} > 0$**: $e^2 > 1 \rightarrow$ **Hipérbola** (órbita abierta/no ligada).

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

@interact(e=(0.0, 0.99, 0.01)) # Evitamos e=1 para no dividir por cero en este modelo
def dibuja_trayectoria(e):
    p = 1
    # ELIMINADO: e=0.7 (esto anulaba el interactivo)

    thetaers = np.linspace(0, 2*np.pi, 200)
    rs = p / (1 + e * np.cos(thetaers))

    xfs = rs * np.cos(thetaers)
    yfs = rs * np.sin(thetaers)

    plt.figure(figsize=(6, 6))
    plt.plot(xfs, yfs, label=f'Excentricidad: {e:.2f}')

    # Dibujar el foco (el Sol o cuerpo central)
    plt.plot(0, 0, 'ro', label="Foco")

    # Ajustamos los límites dinámicamente para que la órbita no se corte
    limit = p / (1 - e) if e < 1 else 10
    plt.xlim(-limit * 1.2, limit * 0.5) # La elipse crece hacia la izquierda
    plt.ylim(-limit, limit)

    plt.axhline(0, color='black', lw=1)
    plt.axvline(0, color='black', lw=1)
    plt.grid(ls="--")
    plt.legend()
    plt.show()
```
