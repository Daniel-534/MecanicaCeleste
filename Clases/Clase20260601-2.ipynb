### Lagrangiano del problema de los dos cuerpos
$$L_{\text{2B,R}} = \frac{1}{2} M \dot{\vec{r}}_{\text{CM}}^2 + \frac{1}{2} m_r \dot{\vec{r}}^2 + \frac{GM m_r}{r}$$


```python
import numpy as np
import sympy as sp
from sympy.physics.mechanics import dynamicsymbols, init_vprinting
import matplotlib.pyplot as plt

init_vprinting()
```

### Lagrangiano del problema de los dos cuerpos
$$L_{\text{2B,R}} = \frac{1}{2} M \dot{\vec{r}}_{\text{CM}}^2 + \frac{1}{2} m_r \dot{\vec{r}}^2 + \frac{GM m_r}{r}$$

```python
M, m_r, G, t = sp.symbols("M m_r G t", positive=True)

x, y, z = dynamicsymbols("x y z")
x_dot, y_dot, z_dot = x.diff(), y.diff(), z.diff()

x_cm, y_cm, z_cm = dynamicsymbols("x_cm y_cm z_cm")

x_cm_dot, y_cm_dot, z_cm_dot = x_cm.diff(), y_cm.diff(), z_cm.diff()

r = sp.sqrt(x**2 + y**2 + z**2)

L = 1/2 * (M * x_cm_dot**2) + 1/2 * (m_r * x_dot**2) + (G * M * m_r) / r

L
```

### Función de Jacobi
$$h \equiv \sum_{j} \frac{\partial L}{\partial \dot{q}_j} \dot{q}_j - L$$


```python
q_dot = [x_cm_dot, x_dot]

sumatoria = sum(sp.diff(L, j) * j for j in q_dot)

h = sumatoria - L

h
```

### Momentums generalizados
$$\frac{\partial L}{\partial \dot{\vec{r}}} = m_r \dot{\vec{r}}, \quad \frac{\partial L}{\partial \dot{\vec{r}}_{\text{CM}}} = m_r \dot{\vec{r}}_{\text{CM}}$$

```python
for j in q_dot:
    display(sp.diff(L,j))
```
