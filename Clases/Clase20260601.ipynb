$$L = \frac{I_1}{2}(\dot{\theta}^2 + \dot{\phi}^2)\sin^2\theta + \frac{I_3}{2}(\dot{\psi} + \dot{\phi}^2\cos\theta)^2-Mgl\cos\theta$$

```python
import sympy as sp
from sympy.physics.mechanics import dynamicsymbols, init_vprinting

# Para notacion de \dot en las derivadas temporales
init_vprinting()

t = sp.symbols('t')
I_1, I_3, M, g, l = sp.symbols('I_1 I_3 M g l')

# Coordenadas generalizadas
theta, phi, psi = dynamicsymbols('theta phi psi')

# Derivadas de coordenadas generalizadas
theta_dot = sp.diff(theta, t)
phi_dot = sp.diff(phi, t)
psi_dot = sp.diff(psi, t)

# Lagrangiano
L = (I_1/2) * (theta_dot**2 + phi_dot**2) * sp.sin(theta)**2 + \
    (I_3/2) * (psi_dot + phi_dot**2 * sp.cos(theta))**2 - M * g * l * sp.cos(theta)
```

```python
# Momentos angulares

display(sp.diff(L, theta_dot))
display(sp.diff(L, phi_dot))
display(sp.diff(L, psi_dot))
```

```python
# Funcion de Jacobi
h = sp.diff(L,t) - L
h.simplify()
```
