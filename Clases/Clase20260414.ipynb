$$\vec{L}_{\text{total}} = \sum\vec{r}\times \vec{p}$$

Aplicar a dos cuerpos.

$$\vec{L}_{\text{total}} = m_1\vec{r}_1\times \dot{\vec{r}}_1 + m_2\vec{r}_2\times \dot{\vec{r}}_2$$

Aplicar cambio a coordenadas de Jacobi

$$\vec{r}_1 = \vec{R}+\frac{m_2}{M}\vec{r}$$

$$\vec{r}_2 = \vec{R}-\frac{m_1}{M}\vec{r}$$

Para simplificar la expresión del momento angular total $\vec{L}_{\text{total}}$ usando coordenadas de Jacobi, primero debemos notar un pequeño ajuste en la definición de $\vec{r}_2$ para que el vector relativo sea $\vec{r} = \vec{r}_1 - \vec{r}_2$.

Las relaciones correctas son:
1. $\vec{r}_1 = \vec{R} + \frac{m_2}{M}\vec{r}$
2. $\vec{r}_2 = \vec{R} - \frac{m_1}{M}\vec{r}$
3. $\dot{\vec{r}}_1 = \dot{\vec{R}} + \frac{m_2}{M}\dot{\vec{r}}$
4. $\dot{\vec{r}}_2 = \dot{\vec{R}} - \frac{m_1}{M}\dot{\vec{r}}$

Donde $M = m_1 + m_2$ y $\vec{R}$ es el centro de masas.

---

### Paso 1: Sustitución de los términos
Sustituimos las expresiones de posición y velocidad en la ecuación original:

$$\vec{L} = m_1 \left( \vec{R} + \frac{m_2}{M}\vec{r} \right) \times \left( \dot{\vec{R}} + \frac{m_2}{M}\dot{\vec{r}} \right) + m_2 \left( \vec{R} - \frac{m_1}{M}\vec{r} \right) \times \left( \dot{\vec{R}} - \frac{m_1}{M}\dot{\vec{r}} \right)$$

### Paso 2: Desarrollo de los productos cruz
Expandimos cada término aplicando la propiedad distributiva del producto vectorial $(\vec{A} \times \vec{B})$:

**Para el primer cuerpo ($m_1$):**
$$m_1 \left[ \vec{R} \times \dot{\vec{R}} + \vec{R} \times \left(\frac{m_2}{M}\dot{\vec{r}}\right) + \left(\frac{m_2}{M}\vec{r}\right) \times \dot{\vec{R}} + \left(\frac{m_2^2}{M^2}\right) \vec{r} \times \dot{\vec{r}} \right]$$

**Para el segundo cuerpo ($m_2$):**
$$m_2 \left[ \vec{R} \times \dot{\vec{R}} - \vec{R} \times \left(\frac{m_1}{M}\dot{\vec{r}}\right) - \left(\frac{m_1}{M}\vec{r}\right) \times \dot{\vec{R}} + \left(\frac{m_1^2}{M^2}\right) \vec{r} \times \dot{\vec{r}} \right]$$

### Paso 3: Agrupación de términos comunes
Sumamos ambas expresiones y agrupamos por el tipo de producto vectorial:

1. **Término de Centro de Masas ($\vec{R} \times \dot{\vec{R}}$):**
$$(m_1 + m_2) (\vec{R} \times \dot{\vec{R}}) = M(\vec{R} \times \dot{\vec{R}})$$

2. **Términos Cruzados ($\vec{R} \times \dot{\vec{r}}$ y $\vec{r} \times \dot{\vec{R}}$):**
$$\left( m_1 \frac{m_2}{M} - m_2 \frac{m_1}{M} \right) (\vec{R} \times \dot{\vec{r}}) = 0$$
$$\left( m_1 \frac{m_2}{M} - m_2 \frac{m_1}{M} \right) (\vec{r} \times \dot{\vec{R}}) = 0$$
*Nota: Estos términos se anulan exactamente, lo cual es la magia de las coordenadas de Jacobi.*

3. **Término de Movimiento Relativo ($\vec{r} \times \dot{\vec{r}}$):**
$$\left( m_1 \frac{m_2^2}{M^2} + m_2 \frac{m_1^2}{M^2} \right) (\vec{r} \times \dot{\vec{r}})$$

### Paso 4: Simplificación del coeficiente relativo
Factorizamos el coeficiente del tercer término:
$$\frac{m_1 m_2^2 + m_2 m_1^2}{M^2} = \frac{m_1 m_2 (m_2 + m_1)}{M^2} = \frac{m_1 m_2 (M)}{M^2} = \frac{m_1 m_2}{M}$$

Este valor se define como la **masa reducida** $\mu = \frac{m_1 m_2}{m_1 + m_2}$.

---

### Resultado Final Simplificado
La expresión se reduce a la suma del momento angular del centro de masas y el momento angular relativo:

$$\vec{L}_{\text{total}} = M(\vec{R} \times \dot{\vec{R}}) + \mu(\vec{r} \times \dot{\vec{r}})$$



Esta forma es extremadamente útil porque separa el movimiento global del sistema del comportamiento orbital interno.

Esta expresión es bellísima porque nos dice que el movimiento de un sistema complejo puede entenderse como la suma de dos movimientos independientes y mucho más simples.

Para explicarlo, imaginemos el sistema **Tierra-Luna**.

---

### 1. El Término del Centro de Masas: $M(\vec{R} \times \dot{\vec{R}})$
Este término representa el **momento angular "global"** del sistema.

* **$M$ (Masa Total):** Es la suma de la masa de la Tierra y la Luna. Conceptualmente, "comprimimos" ambos cuerpos en un solo punto matemático.
* **$\vec{R}$ y $\dot{\vec{R}}$:** Son la posición y la velocidad del **Centro de Masas (Baricentro)** del sistema respecto a un punto externo (por ejemplo, el Sol).

**Concepto:** Imagina que la Tierra y la Luna están unidas por una barra invisible y lanzas ese conjunto al espacio. Este término ignora que los cuerpos están dando vueltas entre sí; solo le importa cómo se desplaza el "punto de equilibrio" del conjunto en su órbita alrededor del Sol. Es el momento angular de una "super-partícula" de masa $M$.

---

### 2. El Término del Movimiento Relativo: $\mu(\vec{r} \times \dot{\vec{r}})$
Este término representa el **momento angular "interno"** o de rotación mutua.

* **$\mu$ (Masa Reducida):** Es una masa ficticia ($\mu = \frac{m_1 m_2}{M}$). Matemáticamente, permite tratar el problema de dos cuerpos como si fuera uno solo. En el caso Tierra-Luna, $\mu$ es ligeramente menor que la masa de la Luna.
* **$\vec{r}$ y $\dot{\vec{r}}$:** Es el vector posición y velocidad de un cuerpo **visto desde el otro** (distancia relativa).

**Concepto:** Este término describe el "baile" interno. Ignora por completo dónde está el Sol o hacia dónde viaja el sistema. Solo se concentra en cuánta inercia de rotación genera la Luna girando alrededor de la Tierra (y la Tierra bamboleándose un poco debido a la Luna).



---

### Ejemplo Práctico: El sistema Tierra-Luna

Si quisiéramos calcular el momento angular total de la Tierra y la Luna mientras viajan por el espacio, la fórmula nos dice:

1.  **$M(\vec{R} \times \dot{\vec{R}})$:** Calcula el momento angular de un punto situado a unos 4,670 km del centro de la Tierra (el baricentro) mientras este sigue una elipse casi perfecta alrededor del Sol.
2.  **$\mu(\vec{r} \times \dot{\vec{r}})$:** Calcula el momento angular de la Luna orbitando a ese baricentro a una distancia media de 384,400 km.

**¿Por qué es útil?**
Porque en la mayoría de los problemas de astronomía, el primer término (la órbita alrededor del Sol) es constante para escalas de tiempo cortas, por lo que los científicos pueden ignorarlo y concentrarse únicamente en el segundo término para estudiar, por ejemplo, cómo cambia la órbita de la Luna con el tiempo.

---

### Resumen de la Identidad
> La rotación total de un sistema es igual a la rotación de su "cascarón" (centro de masas) más la rotación de sus "engranajes internos" (movimiento relativo).

# Anlálisis del vector excentricidad

Para desarrollar el producto escalar del vector de Runge-Lenz (o vector excentricidad) consigo mismo, utilizaremos las propiedades del producto punto y el producto vectorial. Partimos de:

$$\vec{e} = \frac{\vec{v} \times \vec{h}}{\mu} - \hat{r}$$

Donde:
* $\vec{v}$ es el vector velocidad.
* $\vec{h} = \vec{r} \times \vec{v}$ es el momento angular específico.
* $\mu$ es el parámetro gravitacional estándar ($G M$).
* $\hat{r}$ es el vector unitario de posición ($\vec{r}/r$).

### 1. Expansión del producto escalar
Calculamos $e^2 = \vec{e} \cdot \vec{e}$:

$$e^2 = \left( \frac{\vec{v} \times \vec{h}}{\mu} - \hat{r} \right) \cdot \left( \frac{\vec{v} \times \vec{h}}{\mu} - \hat{r} \right)$$

Al distribuir los términos obtenemos:

$$e^2 = \frac{(\vec{v} \times \vec{h}) \cdot (\vec{v} \times \vec{h})}{\mu^2} - 2 \frac{(\vec{v} \times \vec{h}) \cdot \hat{r}}{\mu} + \hat{r} \cdot \hat{r}$$

Aquí tienes una versión mucho más pulida y estructurada, optimizando la claridad matemática y la fluidez del argumento:

### 2. Desglose y Simplificación de los Componentes

Para resolver $e^2 = \frac{(\vec{v} \times \vec{h}) \cdot (\vec{v} \times \vec{h})}{\mu^2} - 2 \frac{(\vec{v} \times \vec{h}) \cdot \hat{r}}{\mu} + \hat{r} \cdot \hat{r}$, analizamos cada término de forma independiente:

* **Término de Normalización:** Dado que $\hat{r}$ es, por definición, un vector unitario que apunta en la dirección de la posición radial, su producto escalar consigo mismo es la unidad:

$$\hat{r} \cdot \hat{r} = |\hat{r}|^2 = 1$$

* **Término Cuadrático de Velocidad:**
    Utilizando la identidad del producto vectorial $(\vec{A} \times \vec{B}) \cdot (\vec{C} \times \vec{D}) = (\vec{A} \cdot \vec{C})(\vec{B} \cdot \vec{D}) - (\vec{A} \cdot \vec{D})(\vec{B} \cdot \vec{C})$, y aplicándola a $(\vec{v} \times \vec{h}) \cdot (\vec{v} \times \vec{h})$:
    $$(\vec{v} \cdot \vec{v})(\vec{h} \cdot \vec{h}) - (\vec{v} \cdot \vec{h})^2$$
    Como el momento angular $\vec{h}$ es siempre perpendicular al plano del movimiento (y por tanto a la velocidad $\vec{v}$), el término $(\vec{v} \cdot \vec{h})$ se anula, resultando en:
    $$\frac{v^2 h^2}{\mu^2}$$

* **Término de Interacción (Cruzado):**
    Aquí empleamos la propiedad de permutación cíclica del producto triple escalar, $\vec{A} \cdot (\vec{B} \times \vec{C}) = \vec{B} \cdot (\vec{C} \times \vec{A})$. Reordenamos el término cruzado para aislar el momento angular:
    $$(\vec{v} \times \vec{h}) \cdot \hat{r} = \vec{h} \cdot (\hat{r} \times \vec{v})$$
    Sustituyendo $\hat{r}$ por su forma expandida $\frac{\vec{r}}{r}$, observamos que aparece nuevamente la definición de momento angular específico ($\vec{h} = \vec{r} \times \vec{v}$):
    $$\vec{h} \cdot \left( \frac{\vec{r}}{r} \times \vec{v} \right) = \frac{1}{r} [\vec{h} \cdot (\vec{r} \times \vec{v})] = \frac{1}{r} (\vec{h} \cdot \vec{h}) = \frac{h^2}{r}$$



Al reensamblar estos componentes, la expresión de la magnitud del vector de Laplace-Runge-Lenz queda lista para su interpretación energética.

### 3. Sustitución en la expresión principal
Reunimos los resultados obtenidos:

$$e^2 = \frac{v^2 h^2}{\mu^2} - \frac{2 h^2}{\mu r} + 1$$

### 4. Relación con la energía orbital
Podemos factorizar $\frac{h^2}{\mu^2}$ de los primeros dos términos:

$$e^2 = \frac{h^2}{\mu^2} \left( v^2 - \frac{2\mu}{r} \right) + 1$$

Recordemos que la energía orbital específica ($\epsilon$) se define como:
$$\epsilon = \frac{v^2}{2} - \frac{\mu}{r} \implies 2\epsilon = v^2 - \frac{2\mu}{r}$$

Sustituyendo esta identidad en nuestra ecuación:

$$e^2 = \frac{h^2}{\mu^2} (2\epsilon) + 1$$

---

### Resultado Final
La magnitud al cuadrado del vector excentricidad es:

$$e^2 = 1 + \frac{2 \epsilon h^2}{\mu^2}$$



**Significado físico:**
* Si $\epsilon < 0$, entonces $e < 1$ (órbita elíptica).
* Si $\epsilon = 0$, entonces $e = 1$ (órbita parabólica).
* Si $\epsilon > 0$, entonces $e > 1$ (órbita hiperbólica).

El resultado $e^2 = 1 + \frac{2\epsilon h^2}{\mu^2}$ no es solo una ecuación de cierre en un libro de texto; es el puente definitivo donde la **geometría de las secciones cónicas** se rinde ante las **leyes de conservación de la energía y el momento angular**. Esta identidad representa una de las síntesis más elegantes de la mecánica clásica, unificando lo que vemos (la forma) con lo que ocurre (la física).

A continuación, se analiza esta conexión mediante tres pilares fundamentales:

### 1. La Geometría como Destino (Lado Izquierdo)
El término $e$ (la excentricidad) es un parámetro puramente geométrico. Define la "estiramiento" de una curva: si es un círculo perfecto ($e=0$), una elipse ($0 < e < 1$), una parábola ($e=1$) o una hipérbola ($e>1$). En este lado de la ecuación, no hay masas, ni fuerzas, ni velocidades; solo existe la forma de la trayectoria trazada en el espacio.

### 2. La Física como Motor (Lado Derecho)
El lado derecho es un compendio de las constantes de movimiento del problema de los dos cuerpos:
* **$\epsilon$ (Energía específica):** Determina si el sistema está "atado" o es libre.
* **$h$ (Momento angular específico):** Define la "limpieza" del giro y la resistencia a caer hacia el centro de atracción.
* **$\mu$ (Parámetro gravitacional):** Representa la intensidad de la fuente que genera el campo.

Aquí reside la causalidad. La física dicta las condiciones bajo las cuales un cuerpo se mueve, y el resultado matemático nos dice que la combinación de cuánta energía tiene el cuerpo y qué tan rápido gira determinará, inevitablemente, el valor de $e$.

### 3. El Umbral de la Parábola
La conexión más fascinante ocurre en el punto crítico $e=1$. Geométricamente, la parábola es la frontera entre una curva cerrada y una abierta. Físicamente, la ecuación revela que esto ocurre exactamente cuando $\epsilon = 0$.

Si la energía es negativa, el término $\frac{2\epsilon h^2}{\mu^2}$ resta a la unidad, obligando a $e$ a ser menor que 1 (una elipse). Si la energía es positiva, suma, lanzando al cuerpo hacia una trayectoria hiperbólica. La ecuación actúa como un **traductor universal**: toma el estado energético de una partícula y lo proyecta en una figura geométrica de Apolonio.

---

### Conclusión
Este resultado nos enseña que en el universo gravitatorio, **la forma es función**. No existe una órbita caprichosa; cada grado de excentricidad está "pagado" con una cantidad precisa de energía y momento angular. La belleza de $e^2 = 1 + \frac{2\epsilon h^2}{\mu^2}$ radica en que desnuda la realidad física, mostrándonos que la geometría de los cielos es, en última instancia, el rastro visible de la conservación de la energía.

Para analizar analíticamente la derivada temporal del vector de excentricidad (o vector de Runge-Lenz), debemos demostrar que, bajo un potencial central de tipo $\propto 1/r$, este vector es una constante de movimiento ($\dot{\vec{e}} = 0$).

Partimos de la definición del vector:
$$\vec{e} = \frac{\vec{v} \times \vec{h}}{\mu} - \hat{r}$$

### 1. Derivación temporal
Aplicamos la regla de la cadena y la regla del producto:
$$\dot{\vec{e}} = \frac{1}{\mu} \left( \dot{\vec{v}} \times \vec{h} + \vec{v} \times \dot{\vec{h}} \right) - \dot{\hat{r}}$$

### 2. Evaluación de los componentes
Para simplificar esta expresión, recordamos las leyes fundamentales del movimiento bajo una fuerza central:

* **Conservación del momento angular:** En un campo de fuerzas centrales, el torque es cero, por lo tanto $\dot{\vec{h}} = 0$. El segundo término del paréntesis desaparece.
* **Segunda Ley de Newton:** La aceleración es $\dot{\vec{v}} = -\frac{\mu}{r^2} \hat{r}$.
* **Derivada del vector unitario:** Por cinemática rotacional, $\dot{\hat{r}} = \vec{\omega} \times \hat{r}$. Dado que $\vec{h} = \vec{r} \times \vec{v} = r^2 \vec{\omega}$, podemos escribir:

$$\dot{\hat{r}} = \frac{\vec{h} \times \hat{r}}{r^2}$$



### 3. Sustitución en la ecuación principal
Sustituimos $\dot{\vec{v}}$ y $\dot{\hat{r}}$ en la expresión de $\dot{\vec{e}}$:

$$\dot{\vec{e}} = \frac{1}{\mu} \left( \left[ -\frac{\mu}{r^2} \hat{r} \right] \times \vec{h} \right) - \frac{\vec{h} \times \hat{r}}{r^2}$$

### 4. Simplificación algebraica
Extraemos los factores comunes:
$$\dot{\vec{e}} = -\frac{1}{r^2} (\hat{r} \times \vec{h}) - \frac{\vec{h} \times \hat{r}}{r^2}$$

Utilizando la propiedad anticomutativa del producto vectorial, donde $-(\hat{r} \times \vec{h}) = (\vec{h} \times \hat{r})$:

$$\dot{\vec{e}} = \frac{\vec{h} \times \hat{r}}{r^2} - \frac{\vec{h} \times \hat{r}}{r^2} = 0$$

---

### Conclusión del análisis
El resultado $\dot{\vec{e}} = 0$ implica que el vector de excentricidad es una **integral de movimiento**.



**Implicaciones físicas:**
1.  **Estabilidad de la órbita:** La constancia de $\vec{e}$ garantiza que la orientación de la órbita (la línea de los ápsides) permanece fija en el espacio. No hay precesión en el problema de Kepler puro.
2.  **Simetría O(4):** Este resultado revela una simetría oculta en el potencial $1/r$. Mientras que la conservación de $\vec{h}$ se debe a la simetría rotacional (espacio isotrópico), la conservación de $\vec{e}$ es específica de la forma matemática de la gravedad, permitiendo que las órbitas se cierren perfectamente tras una revolución.

Para demostrar que en una elipse la suma de las distancias de cualquier punto a los focos ($k$) es igual al eje mayor ($2a$), utilizaremos la definición geométrica de la elipse y analizaremos sus puntos críticos (los vértices).

### 1. Definición geométrica
Una elipse es el lugar geométrico de los puntos $P$ tales que la suma de sus distancias a dos puntos fijos llamados focos ($F_1$ y $F_2$) es constante. Llamaremos a esta constante $k$:
$$d(P, F_1) + d(P, F_2) = k$$



### 2. Elementos de la elipse
Consideremos una elipse centrada en el origen $(0,0)$ con sus focos en el eje $x$:
* **Focos:** $F_1 = (-c, 0)$ y $F_2 = (c, 0)$.
* **Vértices:** $V_1 = (-a, 0)$ y $V_2 = (a, 0)$.
* **Eje mayor:** Es la distancia entre los vértices, la cual es $2a$.

### 3. La prueba por evaluación en un vértice
Dado que la relación $d(P, F_1) + d(P, F_2) = k$ debe cumplirse para **cualquier** punto de la elipse, debe cumplirse específicamente cuando el punto $P$ es uno de los vértices, por ejemplo, $V_2(a, 0)$.

Calculamos las distancias desde $V_2$ a cada foco:
1.  **Distancia al foco lejano ($F_1$):**
    $$d(V_2, F_1) = a - (-c) = a + c$$
2.  **Distancia al foco cercano ($F_2$):**
    $$d(V_2, F_2) = a - c$$

Sumamos ambas distancias para obtener $k$:
$$k = (a + c) + (a - c)$$



### 4. Simplificación
Al realizar la suma, los términos de la semidistancia focal ($c$) se cancelan mutuamente:
$$k = a + c + a - c$$
$$k = 2a$$

---

### Conclusión
La constante de la elipse no es arbitraria; es precisamente la longitud del **eje mayor**. Esto implica que si estiras una cuerda de longitud $2a$ entre dos chinchetas (focos) y la tensas con un lápiz, la figura resultante tendrá exactamente esa anchura total.

Físicamente, esto se relaciona con lo que vimos anteriormente: en el periapsis y apoapsis de una órbita, la suma de los radios vectores siempre compensa la excentricidad para mantener la estructura cerrada de la elipse.

Para demostrar que $p = b^2/a$ de forma puramente geométrica, utilizaremos el **Teorema de Pitágoras** aplicado a un punto específico de la elipse y la definición de la suma de distancias a los focos.

### 1. Construcción del escenario
Consideremos un punto $P$ situado exactamente sobre el foco $F_2$. Por definición, la distancia vertical de $F_2$ a la elipse es el **semilatus rectum** ($p$).

Tenemos los siguientes puntos:
* $F_1, F_2$: Los focos, separados por una distancia $2c$.
* $P$: El punto en la elipse tal que el segmento $PF_2$ es perpendicular al eje mayor.
* $d(P, F_2) = p$ (esta es nuestra incógnita).
* $d(F_1, F_2) = 2c$.



---

### 2. El Triángulo Rectángulo $PF_2F_1$
Al conectar el punto $P$ con el foco lejano $F_1$, se forma un triángulo rectángulo con vértices en $P$, $F_2$ y $F_1$.
* El cateto base es la distancia entre focos: $2c$.
* El cateto altura es el semilatus rectum: $p$.
* La hipotenusa es la distancia $d(P, F_1)$.

Por el Teorema de Pitágoras:
$$[d(P, F_1)]^2 = p^2 + (2c)^2$$
$$d(P, F_1) = \sqrt{p^2 + 4c^2}$$

### 3. Aplicación de la propiedad de la elipse
Sabemos que para cualquier punto de la elipse, la suma de las distancias a los focos es $2a$. Por lo tanto:
$$d(P, F_2) + d(P, F_1) = 2a$$

Sustituimos $d(P, F_2) = p$ y la expresión de la hipotenusa:
$$p + \sqrt{p^2 + 4c^2} = 2a$$

### 4. Resolución algebraica
Aislamos la raíz y elevamos al cuadrado:
$$\sqrt{p^2 + 4c^2} = 2a - p$$
$$p^2 + 4c^2 = (2a - p)^2$$
$$p^2 + 4c^2 = 4a^2 - 4ap + p^2$$

Cancelamos $p^2$ en ambos lados y despejamos $4ap$:
$$4ap = 4a^2 - 4c^2$$
$$ap = a^2 - c^2$$

### 5. Sustitución de la relación fundamental
Recordamos que en la elipse, la relación entre sus dimensiones es $b^2 = a^2 - c^2$. Sustituimos esto en nuestra ecuación:
$$ap = b^2$$

Finalmente, despejamos $p$:
$$p = \frac{b^2}{a}$$

---

### Conclusión Geométrica
El semilatus rectum es la **media proporcional** entre los semiejes de la elipse en relación al eje mayor. Esta construcción geométrica es muy potente porque no depende de coordenadas cartesianas ($x, y$), sino únicamente de las propiedades intrínsecas de la curva y sus focos.
