```python
import pymcel as pc
import numpy as np
import matplotlib.pyplot as plt
import spiceypy as spy
```

```python
tabla, jd, X = pc.consulta_horizons(
    id = "Apophis",
    location = "@SSB",
    epochs = "2026-04-20"
    )
```

```python
rvec = X[:3]
r = np.linalg.norm(rvec)

vvec = X[3:]
v = np.linalg.norm(vvec)
```

```python
hvec = np.cross(rvec, vvec)
h = np.linalg.norm(hvec)

mu = pc.constantes.mu_sun # Aproximacion
evec = np.cross(vvec, hvec) / mu + rvec/r
e = np.linalg.norm(evec)

nvec = np.cross(np.array([0,0,1]),hvec)
n = np.linalg.norm(nvec)
```

```python
p = h**2/mu

rad = 180/np.pi

I = np.arccos(hvec[2]/h)

# Argument of periapsis (omegap)
cos_omegap = np.dot(nvec, evec) / (n * e)
# Handle potential floating point inaccuracies that push cos_omegap slightly out of [-1, 1]
cos_omegap = np.clip(cos_omegap, -1.0, 1.0)
omegap = np.arccos(cos_omegap)
# Determine the quadrant for omegap (argument of periapsis)
# If evec[2] < 0, omegap is in (pi, 2pi)
if evec[2] < 0:
    omegap = 2 * np.pi - omegap

# Longitude of the ascending node (Omega)
cos_Omega = np.dot(nvec, np.array([1, 0, 0])) / n
# Handle potential floating point inaccuracies
cos_Omega = np.clip(cos_Omega, -1.0, 1.0)
Omegap = np.arccos(cos_Omega)
# Determine the quadrant for Omega (longitude of ascending node)
# If nvec[1] < 0, Omega is in (pi, 2pi)
Omega = Omegap if nvec[1] >= 0 else 2 * np.pi - Omegap


# True anomaly (f)
cos_f = np.dot(evec, rvec) / (e * r)
# Handle potential floating point inaccuracies
cos_f = np.clip(cos_f, -1.0, 1.0)
fp = np.arccos(cos_f)
# Determine the quadrant for f (true anomaly)
# If rvec@vvec (dot product) is negative, the object is moving towards periapsis
# so f is in (pi, 2pi)
if np.dot(rvec, vvec) < 0:
    f = 2 * np.pi - fp
else: # Moving away from periapsis, f is in (0, pi)
    f = fp
print(f*rad)
```
