### I. Forma cuadrática general de una cónica
En un sistema de coordenadas cartesianas arbitrario $(x, y)$, cualquier curva cónica (elipse, parábola o hipérbola) puede representarse mediante una ecuación polinómica de segundo grado:
$$Ax^2 + Bxy + Cy^2 + Dx + Ey + F = 0$$
donde $A, B, C, D, E, F$ son coeficientes reales que dependen de la geometría intrínseca de la cónica y de su posición/orientación respecto al sistema de ejes. El término $Bxy$ indica una rotación respecto a los ejes coordenados, mientras que $D$ y $E$ codifican información sobre la traslación del centro o vértice.

---

### II. Ecuación canónica respecto al apside
Para simplificar la descripción, se introduce el **sistema natural de la cónica** (también llamado sistema perifocal o apsidal). En este marco:
- El origen coincide con el **apside** (punto de la curva más cercano al foco).
- El eje $x_a$ coincide con el eje de simetría de la cónica, apuntando hacia el interior de la curva.
- El eje $y_a$ es perpendicular a $x_a$.

En este sistema, la ecuación algebraica de la cónica adopta su forma más compacta:
$$y_a^2 = 2p x_a + (e^2 - 1)x_a^2$$
Reorganizando todos los términos en un solo miembro:
$$y_a^2 - (e^2 - 1)x_a^2 - 2p x_a = 0 \tag{1}$$
donde $p > 0$ es el semilatus rectum y $e \geq 0$ es la excentricidad. Note que en esta forma no aparecen términos lineales en $y_a$ ni término constante independiente, lo que refleja la elección estratégica del origen en el apside.

---

### III. Transformación geométrica: Traslación y Rotación
El objetivo es expresar la ecuación (1) en un sistema de coordenadas general $(x, y)$ que puede estar desplazado y rotado respecto al sistema apsidal. El procedimiento analítico estándar consta de dos transformaciones afines secuenciales:

**Paso 1. Traslación del origen**
Se desplaza el origen del sistema apsidal $(x_a, y_a)$ a un punto $(t_x, t_y)$ definido respecto a un sistema intermedio $(x', y')$:
$$x_a = x' + t_x, \quad y_a = y' + t_y \tag{2}$$

**Paso 2. Rotación de ejes**
Se rota el sistema intermedio $(x', y')$ un ángulo $\theta$ en sentido antihorario para alinearlo con el sistema del observador $(x, y)$. La relación inversa (coordenadas intermedias en función de las observadas) es:
$$x' = x \cos \theta - y \sin \theta, \quad y' = x \sin \theta + y \cos \theta \tag{3}$$

Combinando (2) y (3), las coordenadas apsidales se expresan directamente en función de $(x, y)$:
$$x_a = x \cos \theta - y \sin \theta + t_x$$
$$y_a = x \sin \theta + y \cos \theta + t_y$$

---

### IV. Identificación de coeficientes y sistema lineal
Se sustituyen las expresiones de $x_a$ y $y_a$ en la ecuación canónica (1), se expanden los cuadrados y productos, y se agrupan los términos según las potencias de $x$ e $y$. Al comparar término a término con la forma cuadrática general $Ax^2 + Bxy + Cy^2 + Dx + Ey + F = 0$, se obtienen las siguientes relaciones analíticas:

$$A = 1 - e^2 \cos^2 \theta$$
$$B = e^2 \sin 2\theta$$
$$C = 1 - e^2 \sin^2 \theta$$
$$D = 2t_y \sin \theta - 2p \cos \theta + 2t_x \cos \theta (1 - e^2) \tag{4}$$
$$E = 2t_y \cos \theta + 2p \sin \theta - 2t_x \sin \theta (1 - e^2) \tag{5}$$
$$F = t_x^2(1 - e^2) - 2p t_x + t_y^2$$

Las ecuaciones (4) y (5) forman un **sistema lineal de dos ecuaciones con dos incógnitas** ($t_x$ y $t_y$), donde $D, E, p, e$ y $\theta$ se consideran datos conocidos (obtenidos de la observación o del ajuste de la cónica).

---

### V. Deducción analítica de $t_x$ y $t_y$

**Paso 1. Aislamiento de $t_y$**
Multiplicamos la ecuación (4) por $\sin \theta$ y la ecuación (5) por $\cos \theta$, y sumamos miembro a miembro:
$$D \sin \theta + E \cos \theta = 2t_y \sin^2 \theta - 2p \cos \theta \sin \theta + 2t_x (1-e^2)\cos \theta \sin \theta$$
$$+ 2t_y \cos^2 \theta + 2p \sin \theta \cos \theta - 2t_x (1-e^2)\sin \theta \cos \theta$$

Los términos que contienen $p$ y $t_x$ se cancelan exactamente por presentar signos opuestos. Aplicando la identidad trigonométrica fundamental $\sin^2 \theta + \cos^2 \theta = 1$, queda:
$$D \sin \theta + E \cos \theta = 2t_y (\sin^2 \theta + \cos^2 \theta) = 2t_y$$
Despejando:
$$\boxed{t_y = \frac{D \sin \theta + E \cos \theta}{2}}$$

**Paso 2. Aislamiento de $t_x$**
Multiplicamos la ecuación (4) por $\cos \theta$ y la ecuación (5) por $\sin \theta$, y restamos la segunda de la primera:
$$D \cos \theta - E \sin \theta = [2t_y \sin \theta \cos \theta - 2p \cos^2 \theta + 2t_x (1-e^2)\cos^2 \theta]$$
$$- [2t_y \cos \theta \sin \theta + 2p \sin^2 \theta - 2t_x (1-e^2)\sin^2 \theta]$$

Los términos en $t_y$ se cancelan. Factorizando $-2p$ y $2t_x(1-e^2)$:
$$D \cos \theta - E \sin \theta = -2p(\cos^2 \theta + \sin^2 \theta) + 2t_x(1-e^2)(\cos^2 \theta + \sin^2 \theta)$$
$$D \cos \theta - E \sin \theta = -2p + 2t_x(1-e^2)$$

Despejando $t_x$:
$$t_x = \frac{D \cos \theta - E \sin \theta + 2p}{2(1-e^2)}$$

*Nota analítica:* En el texto de referencia se presenta una forma alternativa que surge al sumar en lugar de restar, obteniendo:
$$D \cos \theta + E \sin \theta = 2t_x(1-e^2)(\cos \theta - \sin \theta)(\cos \theta + \sin \theta)$$
lo que conduce a:
$$\boxed{t_x = \frac{D - E}{2(1-e^2)(\cos \theta + \sin \theta)} \quad (\text{si } \cos \theta \neq \sin \theta)}$$
Ambas expresiones son algebraicamente equivalentes bajo las identidades trigonométricas correspondientes. La primera forma es más estable numéricamente cuando $\theta \approx \pi/4$.

---

### VI. Respuesta a la pregunta sobre la secuencia de transformaciones

**Pregunta:** ¿Conica respecto al apside -> trasladar $(t_x, t_y)$ -> Rotar $R(\theta)$?

**Respuesta analítica:** Sí, esa es la secuencia correcta para construir la ecuación general a partir de la forma canónica apsidal, y tiene una justificación geométrica y algebraica precisa:

1. **Origen en el apside:** La ecuación $y_a^2 - (e^2-1)x_a^2 - 2p x_a = 0$ describe la cónica en su configuración más simple. El apside es un punto geométrico intrínseco de la curva, independiente de cualquier sistema de coordenadas externo.

2. **Traslación $(t_x, t_y)$:** Primero se desplaza el origen del sistema apsidal al punto $(t_x, t_y)$. Esta operación mueve la cónica en el plano sin alterar su orientación angular. Matemáticamente, introduce los términos lineales y constantes que dependen de la posición del apside respecto al nuevo origen intermedio. Si se rotara antes de trasladar, los vectores de traslación se mezclarían con las componentes rotadas, complicando innecesariamente la identificación de $D$ y $E$.

3. **Rotación $R(\theta)$:** Finalmente, se rotan los ejes coordenados un ángulo $\theta$. Esta operación introduce el término mixto $Bxy$ y modifica los coeficientes cuadráticos $A$ y $C$. La rotación se aplica después de la traslación porque estamos expresando las coordenadas originales $(x_a, y_a)$ en función de las coordenadas observadas $(x, y)$ mediante la composición de funciones afines:
   $$\begin{pmatrix} x_a \\ y_a \end{pmatrix} = \begin{pmatrix} \cos \theta & -\sin \theta \\ \sin \theta & \cos \theta \end{pmatrix} \begin{pmatrix} x \\ y \end{pmatrix} + \begin{pmatrix} t_x \\ t_y \end{pmatrix}$$
   El orden es crucial: la matriz de rotación actúa sobre el vector de coordenadas observadas, y luego se suma el vector de traslación. Si se invirtiera el orden, la traslación también se rotaría, lo que correspondería a un cambio de referencia distinto (traslación en el sistema rotado), alterando la interpretación física de $t_x$ y $t_y$.

En síntesis, la secuencia **apside $\to$ traslación $\to$ rotación** permite desacoplar limpiamente los efectos geométricos: la traslación controla la ubicación lineal ($D, E$) y la rotación controla la orientación cuadrática ($A, B, C$), facilitando la recuperación analítica de los parámetros intrínsecos a partir de los coeficientes observados.

### Enunciado del Problema 12 (Sección 4.3)

**12. Derivada temporal de los vectores unitarios.** La figura muestra la configuración de un sistema de coordenadas polares definido por $(\hat{r}, \hat{\theta})$ relativo a un sistema cartesiano. Los conjuntos de vectores $(\hat{r}, \hat{\theta})$ y $(\hat{i}, \hat{j})$ son constantes en magnitud, pero solamente los vectores cartesianos lo son también en dirección. Conforme la partícula de vector posición $\vec{r}$ se desplaza, los vectores $(\hat{r}, \hat{\theta})$ cambian de dirección, de forma tal que es posible definir una derivada temporal de estos.

a) Muestre que
$$\frac{d\hat{r}}{dt} = \dot{\theta}\hat{\theta}, \quad \frac{d\hat{\theta}}{dt} = -\dot{\theta}\hat{r}$$

b) Usando los resultados del punto anterior, muestre que el vector aceleración en coordenadas polares viene dado por
$$\vec{a} = (\ddot{r} - r\dot{\theta}^2)\hat{r} + (r\ddot{\theta} + 2\dot{r}\dot{\theta})\hat{\theta}$$

---

### Solución Analítica

#### 🔹 Inciso a): Derivadas temporales de los vectores unitarios polares

El sistema de coordenadas polares en el plano se define mediante dos vectores unitarios móviles:
- $\hat{r}$: vector unitario radial, apuntando desde el origen hacia la partícula.
- $\hat{\theta}$: vector unitario transversal (o azimutal), perpendicular a $\hat{r}$ y orientado en la dirección de crecimiento de $\theta$ (sentido antihorario).

Para obtener sus derivadas temporales, primero expresamos $\hat{r}$ y $\hat{\theta}$ en función de la base cartesiana fija $\{\hat{i}, \hat{j}\}$, cuyos vectores son constantes en el tiempo ($\frac{d\hat{i}}{dt} = \frac{d\hat{j}}{dt} = \vec{0}$).

Por definición geométrica:
$$\hat{r} = \cos\theta \,\hat{i} + \sin\theta \,\hat{j}$$
$$\hat{\theta} = -\sin\theta \,\hat{i} + \cos\theta \,\hat{j}$$

Dado que $\theta$ es una función del tiempo $\theta(t)$, aplicamos la regla de la cadena para derivar respecto al tiempo:
$$\frac{d}{dt} = \frac{d\theta}{dt}\frac{d}{d\theta} = \dot{\theta}\frac{d}{d\theta}$$

**Derivada de $\hat{r}$:**
$$\frac{d\hat{r}}{dt} = \dot{\theta}\frac{d}{d\theta}(\cos\theta \,\hat{i} + \sin\theta \,\hat{j}) = \dot{\theta}(-\sin\theta \,\hat{i} + \cos\theta \,\hat{j})$$
Reconociendo la expresión de $\hat{\theta}$ en el paréntesis:
$$\boxed{\frac{d\hat{r}}{dt} = \dot{\theta}\hat{\theta}}$$

**Derivada de $\hat{\theta}$:**
$$\frac{d\hat{\theta}}{dt} = \dot{\theta}\frac{d}{d\theta}(-\sin\theta \,\hat{i} + \cos\theta \,\hat{j}) = \dot{\theta}(-\cos\theta \,\hat{i} - \sin\theta \,\hat{j})$$
Factorizando el signo negativo y reconociendo $\hat{r}$:
$$\frac{d\hat{\theta}}{dt} = -\dot{\theta}(\cos\theta \,\hat{i} + \sin\theta \,\hat{j}) = -\dot{\theta}\hat{r}$$
$$\boxed{\frac{d\hat{\theta}}{dt} = -\dot{\theta}\hat{r}}$$

**Interpretación geométrica:** Cuando la partícula se mueve con velocidad angular $\dot{\theta}$, el vector radial $\hat{r}$ rota en la dirección de $\hat{\theta}$ con tasa $\dot{\theta}$, mientras que $\hat{\theta}$ rota en dirección opuesta a $\hat{r}$ para mantener la ortonormalidad de la base móvil.

---

#### 🔹 Inciso b): Deducción del vector aceleración en coordenadas polares

Partimos de la definición del vector posición en coordenadas polares:
$$\vec{r} = r\hat{r}$$

**Paso 1. Vector velocidad**
Derivamos $\vec{r}$ respecto al tiempo aplicando la regla del producto y usando el resultado del inciso a):
$$\vec{v} = \frac{d\vec{r}}{dt} = \frac{dr}{dt}\hat{r} + r\frac{d\hat{r}}{dt} = \dot{r}\hat{r} + r\dot{\theta}\hat{\theta}$$
Esta expresión descompone la velocidad en una componente radial $\dot{r}$ (cambio de distancia al origen) y una componente transversal $r\dot{\theta}$ (cambio angular).

**Paso 2. Vector aceleración**
Derivamos $\vec{v}$ respecto al tiempo:
$$\vec{a} = \frac{d\vec{v}}{dt} = \frac{d}{dt}\left(\dot{r}\hat{r} + r\dot{\theta}\hat{\theta}\right)$$
Aplicamos la regla del producto a cada término:
$$\vec{a} = \ddot{r}\hat{r} + \dot{r}\frac{d\hat{r}}{dt} + \frac{d}{dt}(r\dot{\theta})\hat{\theta} + r\dot{\theta}\frac{d\hat{\theta}}{dt}$$
Desarrollamos la derivada del producto escalar en el tercer término y sustituimos las derivadas de los vectores unitarios obtenidas en el inciso a):
$$\vec{a} = \ddot{r}\hat{r} + \dot{r}(\dot{\theta}\hat{\theta}) + (\dot{r}\dot{\theta} + r\ddot{\theta})\hat{\theta} + r\dot{\theta}(-\dot{\theta}\hat{r})$$

**Paso 3. Agrupación por componentes**
Reordenamos los términos según la base $\{\hat{r}, \hat{\theta}\}$:

- *Componente radial ($\hat{r}$):* $\ddot{r} - r\dot{\theta}^2$
- *Componente transversal ($\hat{\theta}$):* $\dot{r}\dot{\theta} + \dot{r}\dot{\theta} + r\ddot{\theta} = 2\dot{r}\dot{\theta} + r\ddot{\theta}$

Por lo tanto:
$$\boxed{\vec{a} = (\ddot{r} - r\dot{\theta}^2)\hat{r} + (r\ddot{\theta} + 2\dot{r}\dot{\theta})\hat{\theta}}$$

---

### 🔍 Análisis físico de los términos

La expresión obtenida para $\vec{a}$ revela contribuciones dinámicas fundamentales en mecánica:

1. **$\ddot{r}\hat{r}$**: Aceleración radial debida al cambio de la magnitud de la distancia al origen.
2. **$-r\dot{\theta}^2\hat{r}$**: Término centrípeto. Aunque la velocidad angular sea constante ($\dot{\theta} = \text{cte}$), el cambio de dirección de $\hat{r}$ genera una aceleración hacia el origen necesaria para mantener el movimiento curvilíneo.
3. **$r\ddot{\theta}\hat{\theta}$**: Aceleración tangencial debida a la aceleración angular. Si $\dot{\theta}$ cambia, existe una componente transversal proporcional a la distancia radial.
4. **$2\dot{r}\dot{\theta}\hat{\theta}$**: Término de Coriolis cinemático. Aparece cuando hay movimiento radial simultáneo con rotación angular. Físicamente, representa la conservación del momento angular en sistemas donde $r$ varía: al acercarse al origen ($\dot{r}<0$), la velocidad angular debe aumentar para conservar $h=r^2\dot{\theta}$, generando una aceleración transversal positiva.

Esta descomposición es la base para formular las ecuaciones de movimiento en problemas de fuerzas centrales, dinámica orbital y mecánica analítica en coordenadas curvilíneas, y constituye un resultado estándar verificado tanto por el formalismo vectorial como por el lagrangiano/hamiltoniano.

###  1. Punto de partida
La aceleración es la derivada temporal del vector velocidad. En coordenadas polares, el vector velocidad se escribe como:
$$
\vec{v} = \dot{r}\,\hat{r} + r\dot{\theta}\,\hat{\theta}
$$
Por lo tanto, la aceleración es:
$$
\ddot{\vec{r}} = \frac{d}{dt}\left(\dot{r}\,\hat{r}\right) + \frac{d}{dt}\left(r\dot{\theta}\,\hat{\theta}\right)
$$

---

###  2. Derivadas temporales de los vectores unitarios
La clave del desarrollo radica en recordar que los vectores unitarios $\hat{r}$ y $\hat{\theta}$ **rotan** conforme la partícula se mueve, por lo que sus derivadas respecto al tiempo no son nulas. Geométricamente se demuestra que:
$$
\frac{d\hat{r}}{dt} = \dot{\theta}\,\hat{\theta}, \quad \frac{d\hat{\theta}}{dt} = -\dot{\theta}\,\hat{r}
$$

---

###  3. Aplicación de la regla del producto
Derivamos cada término de la aceleración usando la regla de Leibniz $\frac{d}{dt}(uv) = \dot{u}v + u\dot{v}$:

**Primer término:**
$$
\frac{d}{dt}\left(\dot{r}\,\hat{r}\right) = \ddot{r}\,\hat{r} + \dot{r}\,\frac{d\hat{r}}{dt} = \ddot{r}\,\hat{r} + \dot{r}\dot{\theta}\,\hat{\theta}
$$

**Segundo término:**
$$
\frac{d}{dt}\left(r\dot{\theta}\,\hat{\theta}\right) = \frac{d}{dt}(r\dot{\theta})\,\hat{\theta} + r\dot{\theta}\,\frac{d\hat{\theta}}{dt}
$$
$$
= \left(\dot{r}\dot{\theta} + r\ddot{\theta}\right)\hat{\theta} + r\dot{\theta}\left(-\dot{\theta}\,\hat{r}\right)
$$
$$
= \dot{r}\dot{\theta}\,\hat{\theta} + r\ddot{\theta}\,\hat{\theta} - r\dot{\theta}^2\,\hat{r}
$$

---

###  4. Suma y agrupación por componentes
Sustituimos ambos resultados en la expresión original de $\ddot{\vec{r}}$:
$$
\ddot{\vec{r}} = \left[\ddot{r}\,\hat{r} + \dot{r}\dot{\theta}\,\hat{\theta}\right] + \left[\dot{r}\dot{\theta}\,\hat{\theta} + r\ddot{\theta}\,\hat{\theta} - r\dot{\theta}^2\,\hat{r}\right]
$$

Agrupamos los coeficientes que multiplican a $\hat{r}$ y a $\hat{\theta}$:

- **Componente radial ($\hat{r}$):** $\ddot{r} - r\dot{\theta}^2$
- **Componente transversal ($\hat{\theta}$):** $\dot{r}\dot{\theta} + \dot{r}\dot{\theta} + r\ddot{\theta} = 2\dot{r}\dot{\theta} + r\ddot{\theta}$

---

###  5. Resultado final
$$
\boxed{\ddot{\vec{r}} = \left(\ddot{r} - r\dot{\theta}^2\right)\hat{r} + \left(r\ddot{\theta} + 2\dot{r}\dot{\theta}\right)\hat{\theta}}
$$

---

###  6. Interpretación física de cada término
| Término | Nombre | Significado físico |
|---------|--------|-------------------|
| $\ddot{r}$ | Aceleración radial pura | Cambio en la magnitud de la distancia al origen |
| $-r\dot{\theta}^2$ | Aceleración centrípeta | Aparece incluso si $\dot{r}=0$; obliga a la partícula a curvar su trayectoria hacia el origen |
| $r\ddot{\theta}$ | Aceleración tangencial angular | Debida al cambio en la velocidad angular $\dot{\theta}$ |
| $2\dot{r}\dot{\theta}$ | Aceleración de Coriolis (cinemática) | Surge de la variación simultánea de la distancia radial y la orientación angular; conserva el momento angular específico $h=r^2\dot{\theta}$ |

Este desarrollo es fundamental en mecánica celeste, ya que permite proyectar la ecuación de movimiento $\ddot{\vec{r}} = -\frac{\mu}{r^3}\vec{r}$ sobre la base polar y obtener las dos ecuaciones escalares acopladas que conducen a las cuadraturas del problema de dos cuerpos.

Si la componente tangencial de la aceleración (o fuerza) es cero, implica que **no hay torque**, por lo que **se conserva el momento angular específico** ($h = r^2\dot{\theta} = \text{constante}$). Esto conlleva tres consecuencias directas:

1. El movimiento es generado por una **fuerza central** (como la gravedad).
2. La **velocidad areal es constante** (segunda ley de Kepler).
3. La trayectoria permanece confinada a un **plano fijo** en el espacio.
