# Ecuación de Lagrange

El formalismo de Lagrange es increíblemente poderoso porque no se limita a "fuerzas" y "masas"; se trata de encontrar el **camino extremo** (mínimo o máximo) de una cantidad funcional. Aquí tienes cómo se manifiestan estas ecuaciones en otras áreas de la física:

---

## 1. Óptica: El Principio de Fermat

En lugar de minimizar la acción, en óptica minimizamos el **tiempo de viaje** de la luz. Si un rayo de luz viaja en un medio con un índice de refracción variable $n(x, y, z)$, el tiempo total es la integral del camino óptico.

Para un rayo en un plano $x-y$, donde $y = y(x)$, el funcional a minimizar es:


$$S = \int n(x, y) \sqrt{1 + y'^2} \, dx$$

Aquí, el "Lagrangiano" óptico es $L(y, y', x) = n(x, y) \sqrt{1 + (y')^2}$. La ecuación de Euler-Lagrange nos da la trayectoria del rayo:


$$\frac{d}{dx} \left( \frac{n y'}{\sqrt{1 + y'^2}} \right) - \sqrt{1 + y'^2} \frac{\partial n}{\partial y} = 0$$

* **Caso simple:** Si $n$ es constante (el vacío), obtenemos $y'' = 0$, lo que significa que la luz viaja en **línea recta**.
* **Refracción:** De aquí se deriva directamente la **Ley de Snell**.

## 2. Relatividad General: Geodésicas en el Espacio-Tiempo

En el espacio-tiempo curvo, una partícula "libre" sigue el camino más corto (o largo) entre dos puntos, llamado **geodésica**. Aquí, las coordenadas generalizadas son las coordenadas espacio-temporales $x^\mu (\tau)$, donde $\tau$ es el tiempo propio.

El Lagrangiano se define a partir de la métrica $g_{\mu\nu}$:


$$L = \sqrt{g_{\mu\nu} \dot{x}^\mu \dot{x}^\nu}$$

Al aplicar las ecuaciones de Euler-Lagrange $\frac{d}{d\tau} \left( \frac{\partial L}{\partial \dot{x}^\sigma} \right) - \frac{\partial L}{\partial x^\sigma} = 0$, obtenemos la **ecuación de la geodésica**:


$$\frac{d^2 x^\sigma}{d\tau^2} + \Gamma^\sigma_{\mu\nu} \frac{dx^\mu}{d\tau} \frac{dx^\nu}{d\tau} = 0$$

Donde $\Gamma^\sigma_{\mu\nu}$ son los símbolos de Christoffel. Esta ecuación dicta cómo se mueven los planetas alrededor del Sol o cómo se curva la luz cerca de un agujero negro.

## 3. Teoría Cuántica de Campos (QFT)

En física de partículas, no hablamos de la posición de una partícula, sino de **campos** $\phi(x)$ que llenan todo el espacio. En lugar de un Lagrangiano $L$, usamos una **Densidad Lagrangiana** $\mathcal{L}$, que depende del campo y de sus derivadas espacio-temporales $\partial_\mu \phi$.

La ecuación de Euler-Lagrange para campos es:


$$\partial_\mu \left( \frac{\partial \mathcal{L}}{\partial (\partial_\mu \phi)} \right) - \frac{\partial \mathcal{L}}{\partial \phi} = 0$$

### Ejemplo: El Campo de Klein-Gordon

Para una partícula escalar masiva (como el bosón de Higgs), la densidad es:


$$\mathcal{L} = \frac{1}{2} (\partial_\mu \phi)(\partial^\mu \phi) - \frac{1}{2} m^2 \phi^2$$

Al aplicar Euler-Lagrange, obtenemos la ecuación de onda relativista:


$$(\Box + m^2)\phi = 0$$

### Resumen de la analogía

| Campo | Variable | ¿Qué se minimiza/maximiza? |
| --- | --- | --- |
| **Mecánica** | Posición $q(t)$ | Acción Física ($S$) |
| **Óptica** | Trayectoria $y(x)$ | Tiempo de vuelo ($t$) |
| **Relatividad** | Coordenadas $x^\mu$ | Intervalo espacio-temporal ($s$) |
| **Cuántica** | Campo $\phi(x)$ | Acción del campo |
