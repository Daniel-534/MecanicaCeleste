```python
import numpy as np
import sympy as sp
from sympy.physics.mechanics import dynamicsymbols, init_vprinting
import matplotlib.pyplot as plt

init_vprinting()
```

```python
alpha = sp.Symbol("alpha")
x, y, z = dynamicsymbols("x y z")
x_dot, y_dot, z_dot = x.diff(), y.diff(), z.diff()

r1 = sp.sqrt((x+alpha)**2 + y**2 + z**2)
r2 = sp.sqrt((x - 1 + alpha)**2 + y**2 + z**2)

V_mod = -(1-alpha)/r1 - alpha/r2 - 1/2*(x**2+y**2)

L = 1/2 * (x_dot**2 + y_dot**2 + z_dot**2) - x*y_dot + y*x_dot + V_mod

L.diff(x)
```
